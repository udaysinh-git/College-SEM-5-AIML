```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
qc_not = QuantumCircuit(1)
qc_not.x(0)
```

```python
state_initial = Statevector.from_label('0')
state_after_not = state_initial.evolve(qc_not)
print("Initial State |0>:\n",state_initial)
print("After NOT (X gate):\n", state_after_not)
```

```python
fig1 = plot_bloch_multivector(state_after_not)
display(fig1)
```

```python
# AI generated
qc_not_opposite = QuantumCircuit(1)
qc_not_opposite.x(0)

state_initial_opposite = Statevector.from_label('1') # Start with |1>
state_after_not_opposite = state_initial_opposite.evolve(qc_not_opposite)

print("Initial State |1>:\n", state_initial_opposite)
print("After NOT (X gate):\n", state_after_not_opposite)

fig2 = plot_bloch_multivector(state_after_not_opposite)
display(fig2)
```

```python
qc_and = QuantumCircuit(3)
qc_and.ccx(0,1,2)
```

```python
state_input_and = Statevector.from_label("110")
state_after_and = state_input_and.evolve(qc_and)
print("Input State |110> (A=1,B=1, target=0):\n", state_input_and)
print("After AND gate(Toffoli gate):\n", state_after_and)
```

```python
fig2 = plot_bloch_multivector(state_after_and)
display(fig2)
```

```python
for i , amp in enumerate(state_after_and.data):
  print(f"| {i:03b}> {amp}")
```

```python
qc_or = QuantumCircuit(3)
```

```python
qc_or.x(0)
qc_or.x(1)
qc_or.ccx(0,1,2)
qc_or.x(2)
qc_or.x(0)
qc_or.x(1)
```

```python
state_input_or = Statevector.from_label("010")
state_after_or = state_input_or.evolve(qc_or)

print("Input state |010> (A=0,B=1, Target=0) :\n", state_input_or)
print("After OR gate:\n", state_after_or)

fig4 = plot_bloch_multivector(state_after_or)
display(fig4)

for i , amp in enumerate(state_after_or.data):
  print(f"|{i:03b}> : {amp}")
```
